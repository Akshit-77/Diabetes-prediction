## Importing Libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler
```

## Loading Data

```python
features, labels = load_diabetes(return_X_y=True, as_frame= True, scaled = False)
```

## Exploratory data analysis

```python
# plt.figure(figsize = (40,20))
fig, ax = plt.subplots(2,2, figsize = (20,10))
sns.histplot(data = features, x = 'age' ,bins = 20, ax = ax[0,0])
sns.countplot(data = features, x = 'sex' , ax = ax[0,1])
sns.boxplot(data = features, x = 'bmi', ax = ax[1,0])
ax[1,0].axvline(features['bmi'].median(), color = 'r', linewidth = 2, linestyle = '--')
sns.boxplot(data = features, x = 'bp', ax = ax[1,1])
ax[1,1].axvline(features['bp'].median(), color = 'g', linewidth = 2, linestyle = '--')
plt.tight_layout()
plt.show()
```

```python
sns.heatmap(features.corr(), annot=True, cmap='coolwarm', fmt=".2f")
```

```python
sns.pairplot(features)
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, labels, random_state = 42, test_size = 0.2)
```

```python
ss = StandardScaler()
X_train_scaled = ss.fit_transform(X_train.values)
X_test_scaled = ss.transform(X_test.values)
```

```python
print("shape of features is: ", X_train_scaled.shape)
print("shape of target is: ", y_train.shape)
print("type of features is: ", type(X_train_scaled))
print("type of target is: ", type(y_train))
```

```python
print("shape of features is: ", X_test_scaled.shape)
print("shape of target is: ", y_test.shape)
print("type of features is: ", type(X_test_scaled))
print("type of target is: ", type(y_test))
```

## Building Model and Training

```python
rf_regressor = RandomForestRegressor(n_estimators=100, random_state=42)
```

```python
cv_scores = cross_val_score(rf_regressor, X_train_scaled, y_train, cv=10, scoring="neg_mean_squared_error")  
print(f"Cross-Validation Scores: {-cv_scores}")
print(f"Mean CV Accuracy: {np.mean(-cv_scores)}")
print(f"Standard Deviation of CV Accuracy: {np.std(-cv_scores)}")
```

```python
rf_regressor.fit(X_train_scaled, y_train)
```

```python
y_pred = rf_regressor.predict(X_test_scaled)
```

```python
mse = mean_squared_error(y_pred, y_test)
r2 = r2_score(y_test, y_pred)
print("\nMSE of model on test set is: ", mse)
print("\nR2 score of model on test set is: ", r2)
```
